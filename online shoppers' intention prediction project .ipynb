## Handling Outliers

```python
def boxplots(col):
    sns.boxplot(df[col])
    plt.show()

for i in list(df.select_dtypes(exclude=['object','category']).columns)[0:]:
    boxplots(i)
```

variables I have to handle outliers that are:
    
TrafficType , OperatingSystem , ExitRates in %, BounceRates in %, ProductRelated_Duration, ProductRelated,
Admnstrative_Duration, Administrative

### Removing the outliers 

```python
def remove_outliers(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range = Q1 -1.5*IQR
    upper_range = Q3 +1.5*IQR
    return lower_range,upper_range
```

```python
lradminister ,upadminister = remove_outliers(df['Administrative'])
df['Administrative']=np.where(df['Administrative']>upadminister,upadminister,df['Administrative'])
df['Administrative']=np.where(df['Administrative']<lradminister,lradminister,df['Administrative'])

lradministerd ,upadministerd = remove_outliers(df['Administrative_Duration'])
df['Administrative_Duration']=np.where(df['Administrative_Duration']>upadministerd,upadministerd,df['Administrative_Duration'])
df['Administrative_Duration']=np.where(df['Administrative_Duration']<lradministerd,lradministerd,df['Administrative_Duration'])

lrp ,upp = remove_outliers(df['ProductRelated'])
df['ProductRelated']=np.where(df['ProductRelated']>upp,upp,df['ProductRelated'])
df['ProductRelated']=np.where(df['ProductRelated']<lrp,lrp,df['ProductRelated'])

lrpd ,uppd = remove_outliers(df['ProductRelated_Duration'])
df['ProductRelated_Duration']=np.where(df['ProductRelated_Duration']>uppd,uppd,df['ProductRelated_Duration'])
df['ProductRelated_Duration']=np.where(df['ProductRelated_Duration']<lrpd,lrpd,df['ProductRelated_Duration'])

lrbr ,upbr = remove_outliers(df['BounceRates in %'])
df['BounceRates in %']=np.where(df['BounceRates in %']>upbr,upbr,df['BounceRates in %'])
df['BounceRates in %']=np.where(df['BounceRates in %']<lrbr,lrbr,df['BounceRates in %'])

lrer ,uper = remove_outliers(df['ExitRates in %'])
df['ExitRates in %']=np.where(df['ExitRates in %']>uper,uper,df['ExitRates in %'])
df['ExitRates in %']=np.where(df['ExitRates in %']<lrer,lrer,df['ExitRates in %'])

lros ,upos = remove_outliers(df['OperatingSystems'])
df['OperatingSystems']=np.where(df['OperatingSystems']>upos,upos,df['OperatingSystems'])
df['OperatingSystems']=np.where(df['OperatingSystems']<lros,lros,df['OperatingSystems'])

lrtt ,uptt = remove_outliers(df['TrafficType'])
df['TrafficType']=np.where(df['TrafficType']>uptt,uptt,df['TrafficType'])
df['TrafficType']=np.where(df['TrafficType']<lrtt,lrtt,df['TrafficType'])
```

### One hot encoding

```python
df1 = pd.get_dummies(df)
df1.head(20)
```

## Handling Imbalanced dataset

```python
x = df1.drop(['Revenue'],axis=1)#independent variable
y = df1['Revenue']# dependent variable/target variable.
```

```python

sns.countplot(x=y)
```

as above plot shows that target variable 'Revenue' is imbalanced so to balance we use a Randomoversampling technique.

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
x,y = oversample.fit_resample(x,y)
```

```python

sns.countplot(x=y)
```

its balanced perfectly into equal ratio 

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scale  = scaler.fit_transform(x)
```

```python
x_scale
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x_scale,y,test_size = 0.2)
```

we have split the dataset into train and test after completion of preprocessing 

now dataset is ready for modelling , to apply ML algorithms ,,

## Classifcation model

### Decission Tree Classification

```python

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.tree import DecisionTreeClassifier
classifier1 = DecisionTreeClassifier(criterion='entropy', random_state=14)
dt = classifier1.fit(x_train, y_train)
#Prediction
y_pred1 = classifier1.predict(x_test)
# Accuracy
accuracy_DT = accuracy_score(y_test, y_pred1)
accuracy_DT
```

```python
print(classification_report(y_test,y_pred1))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier2 = RandomForestClassifier(n_estimators=500,criterion='entropy', random_state=15)
rfce = classifier2.fit(x_train, y_train)
#Prediction
y_pred2 = classifier2.predict(x_test)
# Accuracy
accuracy_RFCE = accuracy_score(y_test, y_pred2)
accuracy_RFCE
```

```python
print(classification_report(y_test,y_pred2))
```

### RFC with gini

```python
classifier3 = RandomForestClassifier(n_estimators=500,criterion='gini', random_state=15)
rfcg = classifier3.fit(x_train, y_train)
#Prediction
y_pred3 = classifier3.predict(x_test)
# Accuracy
accuracy_RFCG = accuracy_score(y_test, y_pred3)
accuracy_RFCG
```

```python
print(classification_report(y_test,y_pred3))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier4 = LogisticRegression(C=0.4, max_iter=1000,solver='liblinear')
lr = classifier4.fit(x_train, y_train)
#Prediction
y_pred4 = classifier4.predict(x_test)
# Accuracy
accuracy_LR = accuracy_score(y_test, y_pred4)
accuracy_LR
```

```python
print(classification_report(y_test,y_pred4))
```

### Voting Classifier

```python
from sklearn.ensemble import VotingClassifier

evc = VotingClassifier(estimators=[('dt',dt),('rfce',rfce),('rfcg',rfcg),('lr',lr)],
                       voting='hard',flatten_transform=True)
model_evc = evc.fit(x_train,y_train)

#prediction
pred_evc = evc.predict(x_test)
#Accuracy
accuracy_evc = accuracy_score(y_test,pred_evc)

list1 = ['DecissionTree','RandomForest_Entropy','RandomForest_Gini','LogisticRegression','VotingClassifier']
list2 = [accuracy_DT,accuracy_RFCE,accuracy_RFCG,accuracy_LR,accuracy_evc]
list3 = [classifier1,classifier2,classifier3,classifier4,evc]

df_Accuracy = pd.DataFrame({'Method used':list1, 'Accuracy': list2})
print(df_Accuracy)

chart = sns.barplot(x='Method used', y='Accuracy',data=df_Accuracy)  
chart.set_xticklabels(chart.get_xticklabels(),rotation = 90)
print(chart)
```
